```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_brancos = pd.read_csv("Coleta_Brancas-body_petition.csv")
df_pardas = pd.read_csv("Coleta_Pardas-body_petition.csv")
df_negras = pd.read_csv("Coleta_Negras-body_petition.csv")
```

```python
df_geral = pd.concat([df_negras, df_brancos, df_pardas])
```

```python
df_geral = df_geral.drop_duplicates(subset=["Original_Text"], keep="first").reset_index()
df_geral = df_geral.drop(columns="index")
```

```python
df_geral
```

```python
labels = df_geral["Cor"].value_counts().reset_index()
```

# Divisão do DataSet em Relação a Cor

```python
labels
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.pie(labels["count"], labels=labels["Cor"], autopct="%.2f%%", colors=["red", "green", "blue"])
ax.set_title(f"Grupo Étnico Racial da Coleta - Total {df_geral['Conclusao'].count()}")
plt.show()
print(labels)
```

```python
plt.style.use('_mpl-gallery') 
fig, ax = plt.subplots(1,2, figsize=(12,5))
ax[0].pie(labels["count"], labels=labels["Cor"], autopct="%.2f%%", colors=["red", "green", "blue"])
ax[1].bar(labels["Cor"], labels["count"], color=["red", "green", "blue"])
ax[0].set_title(f"Grupo Étnico Racial da Coleta - Total {df_geral['Conclusao'].count()}")
plt.show()
print(df_geral["Cor"].value_counts().reset_index())
```

# Divisão em relação a Conclusão

```python
conclusao = df_geral["Conclusao"].value_counts().reset_index()
```

```python
plt.style.use('_mpl-gallery') 
fig, ax = plt.subplots(1,2, figsize=(12,5))
ax[0].pie(conclusao["count"], labels=conclusao["Conclusao"], autopct="%.2f%%", colors=["red", "green", "blue"])
ax[1].bar(conclusao["Conclusao"], conclusao["count"], color=["red", "green", "blue"])
ax[0].set_title(f"Grupo Étnico Racial da Coleta - Total {df_geral['Conclusao'].count()}")
plt.show()
print(df_geral["Conclusao"].value_counts().reset_index())



```
