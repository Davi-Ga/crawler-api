```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriver
from selenium.webdriver.chromium.service import ChromiumService
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC


options = webdriver.ChromeOptions()
driver = webdriver.Chrome()
wait_component = WebDriverWait(driver, 5)
```

```python
## Componentes
first_result = '//*[@id="rso"]/div[1]/div/div/div[1]'
cor_raca = '/html/body/div[1]/section/div[2]/div[1]/div/div/div[2]/ul/li[3]'
sou_um_robo = '//*[@id="recaptcha-anchor"]/div[1]'
```

```python
import pandas as pd

df = pd.read_csv("cor_da_justica.csv").drop(columns="Unnamed: 0")
```

```python
raca = []
```

```python

for index, name in enumerate(df.get("nome")):   
    try:        
        driver.get(f'https://www.cnnbrasil.com.br/eleicoes/2022/candidatos/deputado-federal/ac/{name.lower()} "CNN"')
        driver.implicitly_wait(1)
        # if wait_component.until(EC.presence_of_element_located((By.XPATH, sou_um_robo))).click:
        #     driver.find_element(By.XPATH, sou_um_robo).click()
            
        wait_component.until(EC.element_located_to_be_selected((By.XPATH, first_result))).click()
        wait_component.until(EC.presence_of_element_located((By.XPATH, cor_raca)))
        cor_racao = driver.find_element(By.XPATH, cor_raca)
        driver.implicitly_wait(1)

        print(cor_racao.text)
        raca.append({"nome": name, "cor": cor_racao.text})
    except:
        raca.append({"nome": name, "cor": "não encontrado"})
        continue
```
