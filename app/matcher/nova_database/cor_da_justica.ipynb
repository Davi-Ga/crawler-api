```python
import pandas as pd
import json

teste = pd.read_json("./data.json")
teste["autor"] = teste["autor"].fillna("placeholder")
teste["autor"] = teste.apply(lambda row: row["relator"] if row["autor"] == "placeholder" else row["autor"], axis=1)
json_object = json.loads('{"ementa": "vazio"}')
# Replace null values in the "proposicao_principal" column with a placeholder value
teste["proposicao_principal"] = teste["proposicao_principal"].fillna("placeholder")
teste["proposicao_principal"] = teste["proposicao_principal"].apply(lambda x: json_object if x == "placeholder" else x)

teste = teste.drop(columns=["relator"])
teste["proposicao"] = teste["proposicao"].drop_duplicates()
```

```python
teste
```

```python

proposicao = pd.json_normalize(teste["proposicao"], max_level=1)
autor = pd.json_normalize(teste["autor"], max_level=1)
proposicao_principal = pd.json_normalize(teste["proposicao_principal"], max_level=1)
```

```python
dataset = pd.concat([ proposicao, autor, proposicao_principal], axis=1)
```
